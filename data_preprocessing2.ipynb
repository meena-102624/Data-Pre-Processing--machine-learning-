#### outliers

```python
retainoutliers = df[['AGE', 'HbA1c', 'BMI']]
```

```python
def find_iqr_outliers(dataframe, column):
    Q1 = dataframe[column].quantile(0.25)
    Q3 = dataframe[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = dataframe[(dataframe[column] < lower_bound) | (dataframe[column] > upper_bound)]
    return outliers
```

```python
for column in retainoutliers.columns:
    outliers = find_iqr_outliers(retainoutliers, column)
    print(f"\nOutliers in {column}:")
    print(outliers)
```

```python
cr = df['Cr'].quantile(0.995)
urea = df['Urea'].quantile(0.999)
filtered_df = df[(df['Cr'] <= cr) & (df['Urea'] <= urea)]
print(f"\n99.5th Percentile Threshold for Cr: {cr}")
print(f"99.9th Percentile Threshold for Urea: {urea}")
```

```python
lipid_columns = ['LDL', 'VLDL', 'HDL', 'TG', 'Chol']
```

```python
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

```python
for column in lipid_columns:
    filtered_df = remove_outliers_iqr(df, column)
    print(f"\nFiltered DataFrame for {column}:")
    print(filtered_df)
```
